```python
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import os

#返回当前进程的工作目录
os.getcwd()
image_raw_data_jpg = tf.gfile.FastGFile('C:/Users/15150/Pictures/Saved Pictures/8.jpg','rb').read()

with tf.Session() as sess:
    img_data = tf.image.decode_jpeg(image_raw_data_jpg)#图像解码
    plt.figure(1)#图像显示
    #print(sess.run(img_data))显示图像矩阵
    plt.imshow(img_data.eval())#对图像处理显示格式
    plt.show()#图像显示
    #为什么加.eval()
    
    #调整图像大小
    resized = tf.image.resize_images(img_data,[400,800],method=0)
    img_data1 = np.asarray(resized.eval(), dtype='uint8')#这里为什么要用这个。
    plt.figure(2)
    plt.imshow(img_data1)
    plt.show
    
    #裁剪和填充图像
    #size大了就裁剪，小了就填充
    croped = tf.image.resize_image_with_crop_or_pad(img_data,600,600)
    padded = tf.image.resize_image_with_crop_or_pad(img_data,2000,2000)
    plt.figure(3)
    plt.imshow(croped.eval())
    plt.show()
    plt.figure(4)
    plt.imshow(padded.eval())
    plt.show()
    
    #对角线翻转图像
    transposed = tf.image.transpose_image(img_data)
    plt.figure()
    plt.imshow(transposed.eval())
    plt.show()
    
    #图像色彩调整
    #在[-max_delta,max_delta]的范围随机调整图片亮度
    adjusted = tf.image.random_brightness(img_data,max_delta = 0.5)
    plt.figure()
    plt.imshow(adjusted.eval())
    plt.show()
    
    #调整图像色调饱和度
    adjusted1 = tf.image.adjust_hue(img_data,0.1)
    plt.figure()
    plt.imshow(adjusted1.eval())
    plt.show()
```
